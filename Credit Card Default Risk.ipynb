```python
# To remove the duplicate values from the dataset
condition = bool(train.duplicated(subset = 'customer_id').any()) 
# only print(train.duplicated()) will suffice to print all the duplicate values
    # only train.drop_duplicates(inplace True) will drop the duplicates

if condition:
    print("There are duplicate customer_ids")
    df = train.drop_duplicates(['customer_id'])
else:
    print("No Duplicate customer_ids")
df = train.drop_duplicates
```

### We cannot delete so many rows as lot of data will be lost

```python
print(train["owns_car"].value_counts()) # No of distint values in a column
print(train["migrant_worker"].value_counts())
```

```python
# Replacing all the null values from the train dataset with mean values or appropriate values

train["owns_car"].fillna("N",inplace=True) # As it is more frequent value
train["no_of_days_employed"].fillna(train["no_of_days_employed"].mean() ,inplace = True)
train["no_of_children"].fillna(train["total_family_members"].mean(),inplace=True)
train["total_family_members"].fillna(train["total_family_members"].mean(),inplace = True)
train["migrant_worker"].fillna(train["migrant_worker"].mean(),inplace=True)
train["yearly_debt_payments"].fillna(train["yearly_debt_payments"].mean(),inplace=True)
train["credit_score"].fillna(train["credit_score"].mean(),inplace=True)

# Replacing all the null values from the test dataset with mean values or appropriate values

test["owns_car"].fillna("N",inplace=True) # As it is more frequent value
test["no_of_days_employed"].fillna(test["no_of_days_employed"].mean() ,inplace = True)
test["no_of_children"].fillna(test["total_family_members"].mean(),inplace=True)
test["total_family_members"].fillna(test["total_family_members"].mean(),inplace = True)
test["migrant_worker"].fillna(test["migrant_worker"].mean(),inplace=True)
test["yearly_debt_payments"].fillna(test["yearly_debt_payments"].mean(),inplace=True)
test["credit_score"].fillna(test["credit_score"].mean(),inplace=True)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression(random_state=1)
logmodel.fit(X_train,y_train)
y_pred = logmodel.predict(X_test)

from sklearn.metrics import  accuracy_score, f1_score, precision_score, recall_score, roc_auc_score
roc=roc_auc_score(y_test, y_pred)

# Accuracy score
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
results = pd.DataFrame([['Logistic Regression', acc,prec,rec, f1,roc]],
columns = ['Model', 'Accuracy', 'Precision', 'Recall', 'F1 Score','ROC'])
results # 0.71
```
